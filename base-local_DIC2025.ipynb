```python
# Importar las dependencias necesarias
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Configuración para Chrome local
op = webdriver.ChromeOptions()

# Crear instancia del navegador
driver = webdriver.Chrome(options=op)

try:
    # Abrir la página web
    driver.get("https://diariooficial.elperuano.pe/Normas")
    
    # Esperar a que los elementos estén disponibles
    wait = WebDriverWait(driver, 10)
    parent_elements = wait.until(
        EC.presence_of_all_elements_located(
            (By.XPATH, "//article[@class='edicionesoficiales_articulos']")
        )
    )
    
    # Guardar resultados en texto.txt
    with open('texto.txt', 'w', encoding='utf-8') as f:
        for parent_element in parent_elements:
            try:
                # Extraer información básica
                titulo = parent_element.find_element(
                    By.XPATH, ".//div[@class='ediciones_texto']/h4"
                ).text.strip()
                
                tipo_norma = parent_element.find_element(
                    By.XPATH, ".//div[@class='ediciones_texto']/h5/a"
                ).text.strip()
                
                descripcion = parent_element.find_element(
                    By.XPATH, ".//div[@class='ediciones_texto']/p[2]"
                ).text.strip()
                
                enlace = parent_element.find_element(
                    By.XPATH, ".//div[@class='ediciones_texto']/h5/a"
                ).get_attribute("href")
                
                # Escribir en formato simple
                f.write(f"{titulo}\n{tipo_norma}\n{descripcion}\n{enlace}\n")
                
            except Exception as e:
                # Si hay error, escribir una línea vacía
                f.write(f"Error al extraer datos\n")
    
    print(f"✅ Se guardaron {len(parent_elements)} artículos en 'texto.txt'")
    
except Exception as e:
    print(f"❌ Error: {e}")
    
finally:
    driver.quit()
```
